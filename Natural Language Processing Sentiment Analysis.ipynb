```python
#removing HTML tags
import re
def remove_tags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r'', text)
```

```python
#remove urls
def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'', text)

df['review'] = df['review'].apply(remove_url)
df.head()
```

```python
def remove_chat_words(text):
    new_text = []
    for i in text.split():
        if i.upper() in chat_words:
            new_text.append(chat_words[i.upper()])
        else:
            new_text.append(i)
    return " ".join(new_text)
```

```python
df['review'] = df['review'].apply(remove_chat_words)
df.head()
```

```python
#emojis 
def remove_emoji(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)


df['review'] = df['review'].apply(remove_emoji)
df.head()
```

<h3> Tokenization</h3>

```python
from nltk.tokenize import word_tokenize,sent_tokenize

df['tokens'] = df['review'].apply(lambda review: word_tokenize(review))
    
#pytanie czy najpierw tokenizacja czy stemming
```

```python
df.head()
```

```python
#stemming
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
def stem_words(text):
    return " ".join([stemmer.stem(word) for word in text.split()])

df['stemming'] = df['review'].apply(lambda review: stem_words(review))
```

```python
df.head()
```

```python
# from nltk.tokenize import word_tokenize,sent_tokenize

df['tokens after stemming'] = df['stemming'].apply(lambda stemming: word_tokenize(stemming))
    
    
#pytanie czy najpierw tokenizacja czy stemming
```

<h3>Naive Bayes Model<h3>

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['stemming'])
y = df['sentiment']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train.shape, X_test.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()
mnb.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = mnb.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

<h3>Logistic Regression</h3>



```python
from sklearn.linear_model import LogisticRegression

# Train the model
lr = LogisticRegression(max_iter=10000)  # Increase max_iter if needed
lr.fit(X_train, y_train)

# Make predictions and evaluate
y_pred = lr.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```
